## ENGENHARIA DE ATRIBUTOS - Criando novos campos com base em data

```python
# Importando os pacotes necess√°rios
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import locale
```

```python
# Carregando o arquivo
df_original = pd.read_csv("dados.csv")
```

```python
df_original.head()
```

```python
#formatando o campo data
locale.setlocale(locale.LC_ALL, 'pt_BR.utf8')

df_original['DATA_VENDA'] =  pd.to_datetime(df_original['DATA_VENDA'], format='%Y/%m/%d %H:%M')
df_original.head()
```

```python
# Criando um campo chamado DATA_INCLUSAO
df_original['DATA_INCLUSAO'] = df_original['DATA_VENDA'].dt.date
df_original.head()
```

```python
# Criando um campo hora e mes de inclus√£o
df_original['HH_MM_INCLUSAO'] = df_original['DATA_VENDA'].dt.strftime('%H:%M')
df_original.head()
```

```python
# Criando um campo somente com a hora
df_original['HORA_INCLUSAO'] = df_original['DATA_VENDA'].dt.strftime('%H')
df_original['HORA_INCLUSAO'] = df_original['HORA_INCLUSAO'].astype(int)
df_original.head()
```

```python
# Criando os campos Ano, Mes e Dia
df_original['ANO_INCLUSAO'] = pd.DatetimeIndex(df_original['DATA_INCLUSAO']).year
df_original['MES_INCLUSAO'] = pd.DatetimeIndex(df_original['DATA_INCLUSAO']).month
df_original['DIA_INCLUSAO'] = pd.DatetimeIndex(df_original['DATA_INCLUSAO']).day
df_original.head()
```

```python
# Criando o campo dia da semana
df_original['DIA_SEMANA_INCLUSAO'] = pd.DatetimeIndex(df_original['DATA_INCLUSAO']).day_name(locale = 'pt_BR.utf8')
df_original.head()
```

```python
# Criando um campo de per√≠odo com uma faixa de acordo com as horas
bins = [-100, 6, 12, 18, 23]
labels = ['MADRUGADA', 'MANHA', 'TARDE', 'NOITE']
df_original['PERIODO_VENDA'] = pd.cut(df_original['HORA_INCLUSAO'], bins=bins, labels=labels)
pd.value_counts(df_original.PERIODO_VENDA)
```

```python
# Visualizando o arquivo
df_original.head()
```

```python
# Gerando o gr√°fico de contagem por dia da semana
plt.rcParams["figure.figsize"] = [14.00, 5.00]
plt.rcParams["figure.autolayout"] = True

sns.countplot(x = 'DIA_SEMANA_INCLUSAO', data = df_original,order=df_original['DIA_SEMANA_INCLUSAO'].value_counts().index)

plt.show()
```

```python
# Gerando o gr√°fico de contagem por periodo
plt.rcParams["figure.figsize"] = [14.00, 5.00]
plt.rcParams["figure.autolayout"] = True

sns.countplot(x = 'PERIODO_VENDA', data = df_original,order=df_original['PERIODO_VENDA'].value_counts().index)

plt.show()
```
