## NERD DOS DADOS - INDICE DE CANCELAMENTO POR FILIAL

```python
import pandas as pd
```

```python
df_original = pd.read_excel("dados.xlsx")
df_original.head(10)
```

```python
df_original.shape
```

```python
df_resumo = df_original.groupby(['FILIAL','SITUACAO']).size().unstack('SITUACAO', fill_value=0)
df_resumo.head(10)
```

```python
df_resumo['TOTAL_CLIENTES_POR_FILIAL'] = df_resumo['ATIVO'] + df_resumo['CANCELADO']
df_resumo.head(10)
```

```python
df_resumo['PERC_CANCELAMENTO'] = df_resumo['CANCELADO'] / df_resumo['TOTAL_CLIENTES_POR_FILIAL'] * 100
df_resumo.head(10)
```

```python
df_resumo.sort_values('PERC_CANCELAMENTO', ascending=False)
```

```python
df_original = pd.read_excel("dados.xlsx")
df_original.head(10)
```
