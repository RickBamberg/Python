```python
import requests
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
# Defina o período desejado
data_inicial = "2023-07-01"
data_final = "2023-07-31"
```

```python
# Construa a URL da API para obter a PTAX
url_ptax = f"https://api.bcb.gov.br/dados/serie/bcdata.sgs.10813/dados?formato=json&dataInicial={data_inicial}&dataFinal={data_final}"
response_ptax = requests.get(url_ptax)
data_ptax = response_ptax.json()

# Extrai as datas e valores da PTAX
dates_ptax = [entry["data"] for entry in data_ptax]
values_ptax = [float(entry["valor"]) for entry in data_ptax]
```

```python
# Obtém os dados do dólar usando a yfinance
dollar_data = yf.download("USDBRL=X", start=data_inicial, end=data_final)

# Calcula os valores máximos e mínimos diários do dólar
max_dollar_values = dollar_data["High"]
min_dollar_values = dollar_data["Low"]
```

```python
# Calcula as variações diárias da PTAX e dos valores máximos e mínimos do dólar
variations_ptax = [values_ptax[i] - values_ptax[i-1] for i in range(1, len(values_ptax))]
variations_max_dollar = [max_dollar_values[i] - max_dollar_values[i-1] for i in range(1, len(max_dollar_values))]
variations_min_dollar = [min_dollar_values[i] - min_dollar_values[i-1] for i in range(1, len(min_dollar_values))]

# Converte as datas para o formato datetime
dates_ptax = [datetime.strptime(date, "%d/%m/%Y") for date in dates_ptax]
```

```python
# Cria o gráfico
plt.figure(figsize=(10, 6))

# Plotar variações diárias da PTAX, Máximo e Mínimo do Dólar
plt.plot(dates_ptax[1:], variations_ptax, color='b', marker='o', label='Variação PTAX')
plt.plot(max_dollar_values.index[1:], variations_max_dollar, color='r', marker='o', label='Variação Máx Dólar')
plt.plot(min_dollar_values.index[1:], variations_min_dollar, color='g', marker='o', label='Variação Mín Dólar')

plt.xlabel('Data')
plt.ylabel('Variação')
plt.title('Variações Diárias da PTAX, Máximo e Mínimo do Dólar')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%d/%m/%Y"))
plt.legend()
plt.tight_layout()

# Exibe o gráfico
plt.show()
```
