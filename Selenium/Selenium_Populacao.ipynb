```python
# Entra no site e extrai as tabelas necessarias
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from unidecode import unidecode
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', None)  # Exibe todas as linhas
pd.set_option('display.max_columns', None)  # Exibe todas as colunas
pd.set_option('display.max_seq_items', None)  # Exibe todos os itens em uma sequência (como uma lista)

# Ativa o Selenium
selenium_service = Service('C:/Users/Carlos/anaconda3/chromedriver')

chrome_options = Options()
chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36")
chrome_options.add_argument('--headless') # Executar em modo headless (sem interface gráfica)

navegador = webdriver.Chrome(service=selenium_service, options=chrome_options)

# Site de pesquisa
url = 'https://pt.m.wikipedia.org/wiki/Lista_de_estados_brasileiros_por_n%C3%BAmero_de_munic%C3%ADpios'
navegador.get(url)

# Aguarde o carregamento completo da página, se necessário
# ...

# Extraia o conteúdo necessário usando o Selenium
html_content = navegador.page_source

# Use o Pandas para ler as tabelas HTML
dfs = pd.read_html(html_content)

# Acesse os DataFrames extraídos conforme necessário
for df in dfs:
    print(df)

navegador.quit()
```

```python
# Transforma as listas em DataFrame
df0 = pd.DataFrame(dfs[0])
display(df0)
df1 = pd.DataFrame(dfs[1])
display(df1)
df2 = pd.DataFrame(dfs[2])
display(df2)
```

```python
display(df0.info())
```

```python
# Retira os caracteres da lingua poutugesa
df0.columns = [unidecode(col) for col in df0.columns]
```

```python
# Trata os Campos
df0['Numero total de municipios.1'] = df0['Numero total de municipios.1'].str.replace(' ', '').astype(int)
df0['Numero total de municipios.1'] = df0['Numero total de municipios.1'].astype('int64')
```

```python
display(df0.info())
```

```python
display(df1.info())
```

```python
# Retira os caracteres da lingua poutugesa
df1.columns = [unidecode(col) for col in df1.columns]
```

```python
# Trata os Campos
df1['Municipios'] = df1['Municipios'].str.replace('\s', '', regex=True).str.replace('\W', '', regex=True).astype(int)
df1['Municipios'] = df1['Municipios'].astype('int64')

df1['Porcentagem'] = df1['Porcentagem'].str.replace('%', '')
df1['Porcentagem'] = df1['Porcentagem'].str.replace('.', '').str.replace(',', '.').astype(float)
```

```python
display(df1.info())
```

```python
# Retira os caracteres da lingua poutugesa
df2.columns = [unidecode(col) for col in df2.columns]
```

```python
# Trata os Campos

df2.rename(columns=lambda x: x.replace('[9]', ''), inplace=True)
df2.rename(columns=lambda x: x.replace('[10]', ''), inplace=True)

df2['Numero de habitantes por estado federado'] = (df2['Numero de habitantes por estado federado']
                                                   .str.replace('\s', '', regex=True).str.replace('\W', '', regex=True)
                                                   .astype(int))
df2['Numero de habitantes por estado federado'] = df2['Numero de habitantes por estado federado'].astype('int64')

df2['Media de habitantes por municipio'] = (df2['Media de habitantes por municipio']
                                            .str.replace('\s', '', regex=True).str.replace(',', '.').astype(float))
```

```python
#Quantidade de estados em cada região
cores = ['#7FFFD4', '#FA8072', '#00BFFF', '#FFDEAD', '#DDA0DD']

plt.pie(df1['Unidades federativas'], labels=df1['Porcentagem'],shadow=True, \
        explode=(0.05,0.05,0.05,0.05,0.05), colors=cores)
plt.legend(df1['Regiao'], title='Região', bbox_to_anchor=(1, 1))
plt.title('Quantidades de estados por região', weight = 'bold', size=15)
plt.show()
```

```python
#Número de municipios por região

plt.subplots(figsize=(7, 5))
sns.barplot(x=df1['Regiao'], y=df1['Municipios'], palette='Set2')

plt.title('Número de municípios por regiões', weight = 'bold', size=15)
plt.xlabel('Região')
plt.ylabel('Número de municípios')
```
