```python
# Passo 1 - Importar Biblioteca
import pandas as pd
import warnings
#warnings.filterwarnings('ignore')
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

```python
# Passo 2 - Configurar o Selenium
selenium_service = Service('C:/Users/Carlos/anaconda3/chromedriver')
chrome_options = Options()
chrome_options.add_argument('--headless') # Executar em modo headless (sem interface grÃ¡fica)
driver = webdriver.Chrome(service=selenium_service, options=chrome_options)
```

```python
# Passo 3 - Extrair os dados da pÃ¡gina
driver.get('https://br.investing.com/currencies/streaming-forex-rates-majors')

# Obter o HTML da pÃ¡gina
html = driver.page_source

# Fechar o navegador
driver.quit()
```

```python
html
```

```python
# passo 4 - Analisar o HTML com BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Localizar a tabela na pÃ¡gina (ajuste de acordo com a estrutura HTML)
tabela = soup.find('table')

# Extrair os dados da tabela
df = pd.DataFrame()
dados = []
linhas = tabela.find_all('tr')
for linha in linhas:
    colunas = linha.find_all('td')
    linha_dados = []
    for coluna in colunas:
        linha_dados.append(coluna.text.strip())
    dados.append(linha_dados)
    print(linha_dados)
df = pd.DataFrame(dados, columns=['CÃ³digo', 'Compra', 'Venda', 'MÃ¡xima', 'MÃ­nima', 'VariaÃ§Ã£o', 'Var. %', 'Hora'])
```

```python
# Eliminar a coluna de indice
df = df.reset_index(drop=True)
```

```python
# Passo 5 - Processando os dados da tabela
```

```python
# Eliminando a Linha individualmente //  axis=0 - Linha // axis=1 - Coluna 
df = df.drop(0,axis=0)
```

```python
# Ajustando as variaves string para os seus devidos tipos

df['Compra'] = df['Compra'].str.replace('.', '').str.replace(',', '.').astype(float)
df['Venda'] = df['Venda'].str.replace('.', '').str.replace(',', '.').astype(float)
df['MÃ¡xima'] = df['MÃ¡xima'].str.replace('.', '').str.replace(',', '.').astype(float)
df['MÃ­nima'] = df['MÃ­nima'].str.replace('.', '').str.replace(',', '.').astype(float)
df['VariaÃ§Ã£o'] = df['VariaÃ§Ã£o'].str.replace('.', '').str.replace(',', '.').astype(float)

df['Var. %'] = df['Var. %'].str.replace('%', '')
df['Var. %'] = df['Var. %'].str.replace('.', '').str.replace(',', '.').astype(float)

df['Hora'] = pd.to_datetime(df['Hora'], format='%H:%M:%S').dt.time

df['CÃ³digo'] = df['CÃ³digo'].str.replace('Layer 1', '')
```
