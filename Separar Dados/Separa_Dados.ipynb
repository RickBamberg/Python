```python
# - Separar Dados (Train, Test)
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Carregar seu dataframe (substitua 'seu_dataframe.csv' pelo caminho do seu arquivo CSV)
df = pd.read_excel("dados.xlsx")
```

```python
# Definir a proporção de treinamento (por exemplo, 80%) e teste (por exemplo, 20%)
percent_treino = 0.8

# Dividir o dataframe em conjuntos de treinamento e teste - Aleatorio
df_treino, df_teste = train_test_split(df, train_size=percent_treino, random_state=42)
```

```python
# Agora você tem o dataframe de treinamento (df_treino) e o dataframe de teste (df_teste)
# Se você quiser dividir o dataframe igualmente em vez de usar uma proporção, pode fazer algo assim:

# Obter o número de linhas do dataframe
num_linhas = len(df)

# Definir o número de linhas para cada conjunto (por exemplo, 50% para cada)
num_linhas_treino = int(num_linhas * 0.5)

# Embaralhar o dataframe para garantir que as linhas sejam selecionadas aleatoriamente
df = df.sample(frac=1, random_state=42)

# Dividir o dataframe igualmente
df_treino = df.iloc[:num_linhas_treino]
df_teste = df.iloc[num_linhas_treino:]
```
