```python
#Pandas: Possui inúmeras funções e comandos para importar arquivos, analisar dados, tratar dados, etc.
import pandas as pd

#Warnings: Possui detalhes sobre os avisos e alertas que aparecem, porém podemos utiliza-lo também para que os alertas de
#futuras atualizações e metodos depreciados não sejam exibidos
import warnings
warnings.filterwarnings("ignore") 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Um DataFrame do Pandas é como se fosse uma planilha do Excel, onde podemos tratar linhas e colunas.
nome = 'Nasdaq.xlsx'
df_dados = pd.read_excel(nome)
```

```python
#Comando utilizado para verificar a quantidade de linhas e colunas do arquivo
#Colunas também são chamadas de variáveis.
df_dados.shape
```

```python
#Comando utilizado para verificar as linhas iniciais do DataFrame
df_dados.head()
```

```python
df = df_dados
```

```python
# Exclui a coluna "Unnamed: 0"
df = df.drop(columns='Unnamed: 0')
```

```python
# exclui as linhas dos acumuladosres
df = df.drop(index=1)
df = df.drop(index=2)
df = df.drop(index=3)
df = df.drop(index=4)
df = df.drop(index=5)
```

```python
# substituir todos os valores null por 0
df.fillna(0, inplace=True)
```

```python
df
```

```python
# Renomear as colunas principais para garantir consistência
df.rename(columns={df.columns[0]: "Data", df.columns[1]: "Semana"}, inplace=True)

# Criar uma lista para armazenar os dados corrigidos
dados_corrigidos = []

# Iterar sobre cada linha da tabela original
for _, row in df.iterrows():
    data = row.iloc[0]  # Pegando a data (primeira coluna)
    semana = row.iloc[1]  # Pegando a semana (segunda coluna)
    
    # Iterar sobre as colunas de ativos (começando da terceira coluna)
    for i in range(2, len(row), 2):  # Pegando os dados de 2 em 2 colunas (Valor, Percentual)
        if i + 1 < len(row):  # Garantir que existe um percentual correspondente
            ativo = df.columns[i]  # Pegando o nome do ativo
            valor = row.iloc[i]  # Pegando o valor
            percentual = row.iloc[i + 1]  # Pegando o percentual
            
            # Verifica se existe movimento (valor diferente de 0 ou NaN)
            if valor != 0 and pd.notna(valor):
                dados_corrigidos.append([data, semana, ativo, valor, percentual])

# Criar um novo DataFrame com os dados corrigidos
df_corrigido = pd.DataFrame(dados_corrigidos, columns=["Data", "Semana", "Ativo", "Valor", "Percentual"])

# Salvar e exibir
df_corrigido.to_excel("dados_transformados.xlsx", index=False)
print(df_corrigido.head())  # Conferir os primeiros resultados
```
