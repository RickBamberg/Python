```python
#Pandas: Possui inúmeras funções e comandos para importar arquivos, analisar dados, tratar dados, etc.
import pandas as pd

#Warnings: Possui detalhes sobre os avisos e alertas que aparecem, porém podemos utiliza-lo também para que os alertas de
#futuras atualizações e metodos depreciados não sejam exibidos
import warnings
warnings.filterwarnings("ignore") 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def transformar_tabela(df):
    """
    Transforma a tabela no formato desejado e formata o percentual.

    Args:
        df: DataFrame original com os dados.

    Returns:
        DataFrame no novo formato.
    """

    # Lista de ativos (você pode ajustar conforme necessário)
    ativos = ['AAPL', 'MSFT', 'NVDA', 'AMZN', 'META', 'GOOG', 'GOOGL', 'TSLA', 'INTC', 'AMD', 'US 2Y', 'US 5Y', 'US 10Y', 'US 30Y', 'VIX', 'DXY', 'WTI', 'Min. Ferro', 'Ouro', 'Milho', 'S&P 500', 'NASDAQ', 'Dow Jones', 'IBOV', 'INDFUT', 'DOLFUT', 'USD/BRL', 'EUR/BRL', 'CAD/BRL', 'USD/CAD', 'USD/EUR', 'Bitcoin']

    # Lista para armazenar os dados transformados
    data = []

    # Itera sobre as linhas do DataFrame
    for i in range(1, len(df)):  # Começa da segunda linha (índice 1)
        data_row = df.iloc[i]
        data_value = data_row['Unnamed: 1']
        semana = data_row['Unnamed: 2']

        # Itera sobre os ativos
        for j, ativo in enumerate(ativos):
            # Calcula os índices das colunas de valor e percentual
            valor_col_index = 3 + (j * 2)
            percentual_col_index = 4 + (j * 2)

            # Para "Invest. Não Residentes
            #if valor_col_index == 67:
            #    valor_col_index = valor_index + 1
            #    percentual_col_index = percentual_col_index + 1

            # Extrai valor e percentual
            valor = data_row[f'Unnamed: {valor_col_index}']
            percentual = data_row[f'Unnamed: {percentual_col_index}']

            # Formata o percentual (multiplica por 100)
            percentual_formatado = percentual * 100

            # Adiciona os dados à lista
            data.append([data_value, semana, ativo, valor, percentual_formatado])

    # Cria o novo DataFrame
    novo_df = pd.DataFrame(data, columns=['data', 'semana', 'ativo', 'Valor', 'Percentual'])

    return novo_df
```

```python
#Um DataFrame do Pandas é como se fosse uma planilha do Excel, onde podemos tratar linhas e colunas.
nome = 'Nasdaq.xlsx'
df_dados = pd.read_excel(nome)
```

```python
#Comando utilizado para verificar as linhas iniciais do DataFrame
df_dados.head()
```

```python
df = df_dados
```

```python
# Exclui a coluna "Unnamed: 0"
df = df.drop(columns='Unnamed: 0')
```

```python
# exclui as linhas dos acumuladosres
df = df.drop(index=1)
df = df.drop(index=2)
df = df.drop(index=3)
df = df.drop(index=4)
df = df.drop(index=5)
```

```python
# substituir todos os valores null por 0
df.fillna(0, inplace=True)
```

```python
#Comando utilizado para verificar as linhas iniciais do DataFrame
df.head()
```

```python
# Aplica a transformação
novo_df = transformar_tabela(df)
```
